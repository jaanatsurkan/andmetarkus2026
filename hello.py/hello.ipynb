## if/else

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C")
        if year < -10000:
            print("Stone age...")
    elif year == 0:
        print("Happy Birthday, Jesus!")

    elif year == 1978:
        print("Happy Birthday, Mati!")
    else:
        print("A.D.")

make_sense_of_this_year(year=-900000)
```

```python
def print_year(year: int):
    if year < 0:
        print(f"{abs(year)} B.C")
    else:
        print(f"{year} A.D.")

print_year(-90)
print_year(2025)
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['target'] = iris_data.target
iris ['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)": "sepal_length",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_length",
    "petal width (cm)": "petal_width"
})
iris
```

```python
iris.to_csv("iris.csv", index=False)
```

```python
duckdb.sql("""
    SELECT 
           species,
           avg(petal_length) AS avg_petal_length
    FROM iris
    GROUP BY species
       ORDER BY avg_petal_length DESC
 """).df()
```

```python
def get_price(hinnad):
    total = sum(hinnad)

    if total > 100:
        total *= 0.8
    elif total > 50:
        total *= 0.9

    return int(total)


hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]

print(get_price(hinnad))
```

```python


def get_discount(price: int) -> float:
    if price > 100:
        return price * 0.8
    if price > 50:
        return price *0.9
    return price

def get_price(prices: list[int]) -> float:
    tulemus = 0
    for hind in prices:
       (allahindlusega_hind) = get_discount(hind)
       tulemus = tulemus + allahindlusega_hind
    return tulemus 

def get_price_short(price: list[int]) -> float:
    return sum([get_discount(hind) for hind in price])

def get_price_short(price: list[int]) -> float:
    print (sum([get_discount(hind) for hind in price]))

hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]
get_price(hinnad)
```
